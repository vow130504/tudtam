#### Giải thuật phân rã ma trận A     =     QR (Gram – Schmidt)

- Bước 1: Giả sử một ma trận $A$ có n cột, mỗi cột của ma trận lần lượt là các vector tương ứng từ $u_1,u_2,...u_n$
- Bước 2: Đặt $v_1      =      u_1$, nếu $v_1$ là vector 0 thì họ vector phụ thuộc tuyến tính và kết thúc giải thuật.
- Bước 3: Đặt $v_2,v_3,...v_n$ lần lượt với $v_m = u_m − \sum_{i=1}^{m-1}\frac{⟨u_m, v_i⟩}{∥v_i∥^{2}}v_i$ (2 <= m <= n). Nếu có bất kì vector $v_m$ nào là vector 0 thì họ vector phụ thuộc tuyến tính và kết thúc giải thuật. (∥v_i∥ là độ dài của vector).
- Bước 4: xây dựng họ các vector trực chuẩn với $q_m = \frac{v_m}{∥v_m∥}$ (1 <= m <= n). 
- Bước 5: Dựng ma trận $Q$ gồm n cột, $Q(q_l,q_2,...,q_n)$
- Bước 6: Dựng ma trận R là một ma trận tam giác trên
* Với
R = $\begin{bmatrix} ⟨u_1, q_l⟩ & ⟨u_2, q_l⟩ & ... & ⟨u_n, q_l⟩ \\
    0 & ⟨u_2, q_2⟩ & ... & ⟨u_n, q_2⟩ \\
    \vdots &\vdots & \ddots &\vdots\\               
    0 & 0 & ... & ⟨u_n, q_n⟩ \end{bmatrix}$

```python
def float_to_fraction(x):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    while abs(decimal - num/denom) > saiSo:
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
    return f"{sign * num}/{denom}"
    

def subtractVector(A,B):
    if(len(A) != len(B)):
        return None
    C = [] 
    for i in range(len(A)):
        C.append(A[i] - B[i])
    return C


def multiVector(c,A):
    D = []
    for i in range(len(A)):
        D.append(c * A[i])
    return D


def printMatrix(A):
    if A is None:
        print('\n')
    else:
        for row in A:
            print([float_to_fraction(elem) for elem in row])
        print('\n')


def multiMatrix(A,B):
    if(len(A[0])!=len(B)):
        print("Không thể nhân 2 ma trận\n")
        return None
    C = []
    ans = 0
    for i in range(len(A)):
        tam = []
        for j in range(len(A[0])):
            ans = 0
            for k in range(len(A[0])):
                ans += A[i][k] * B[k][j]
            tam.append(round(ans,3))
        C.append(tam)
    return C



def dotProduct(A,B):
    if(len(A) != len(B)):
        return 0
    tam = 0
    for i in range(len(A)):
        tam += A[i]*B[i]
    return tam




def isVector0(A):
    for i in range(len(A)):
        if A[i] != 0:
            return False
    return True



def norm(A):
    tam = 0.0
    for i in range(len(A)):
        tam += A[i]**2
        
    return tam**(0.5)



def QRdecomposition(A):
    T = []
    tam = []
    for i in range(len(A[0])):
        for j in range(len(A)):
            tam.append(A[j][i]*1);
        T.append(tam)
        tam = []
        
    Q = []
    tam2 = []
    for i in range(len(T[0])):
        tam2.append(T[0][i]*1)
    Q.append(tam2)
    tam2 = []
    
    for i in range(1,len(T)):
        tam2 = T[i]
        for j in range(0,len(Q)):
            tam = multiVector(dotProduct(T[i],Q[j])/(norm(Q[j])**2),Q[j])
            tam2 = subtractVector(tam2,tam)
        if isVector0(tam2) == True:
            print("Hệ vector phụ thuộc tuyến tính nên không là họ trực giao")
            return None,None
        Q.append(tam2)
    
    #tinh truc chuan    
    for i in range(len(Q)):
        chuan1 = norm(Q[i])
        for j in range(len(Q[0])):
            Q[i][j] = Q[i][j]/chuan1
    
    R = []
    for i in range(len(Q)):
        tam = []
        for j in range(len(T)):         
            if j<i:
                tam.append(0)
                continue
            tam.append(dotProduct(Q[i],T[j]))
        R.append(tam)
        
    Qsau = []
    tam = []
    for i in range(len(Q[0])):
        for j in range(len(Q)):
            tam.append(Q[j][i]);
        Qsau.append(tam)
        tam = []
        
    return Qsau,R
  

def printResolve(A):
    if A == None: 
        return
    print("Ma tran A: ")
    printMatrix(A)
    print('\n')

    Q,R = QRdecomposition(A)
    if Q == None or R == None:
        return
    
    print("Ma trận Q: ")
    printMatrix(Q)
    print('\n')

    print("Ma trận R: ")
    printMatrix(R)
    print('\n')

    C=multiMatrix(Q,R)
    print("Ma trận QxR: ")
    printMatrix(C)
    print('\n')
    
A=[[1,1,-1],[0,1,2],[1,1,1]] #phân rã QR được 
A1=[[1,0,1],[-1,1,1],[1,0,1],[-1,1,1]] #Không là họ trực giao -> không thể phân rã QR
A2=[[-2,1,3],[1,0,0],[0,1,0],[0,0,1]] #phân rã QR được 

print("\n---------Test 1---------")
printResolve(A)
print("\n---------Test 2---------")
printResolve(A1)
print("\n---------Test 3---------")
printResolve(A2)
```

```python
import numpy as np
np.set_printoptions(suppress=True)

def float_to_fraction(x):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    while abs(decimal - num/denom) > saiSo:
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
    return f"{sign * num}/{denom}"
    
def printMatrix(A):
    if A is None:
        print('\n')
    else:
        for row in A:
            print([float_to_fraction(elem) for elem in row])
        print('\n')



def printResolve(A):
    if A is None: 
        return
    
    print("Ma trận A: ")
    printMatrix(A)
    print('\n')

    if np.linalg.matrix_rank(A) < min(len(A), len(A[0])):
        print("Không thể phân rã QR.")
        return

    try:
        Q, R = np.linalg.qr(A)
        print("Ma trận Q: ")
        printMatrix(Q)
        print('\n')

        print("Ma trận R: ")
        printMatrix(R)
        print('\n')

        C = np.dot(Q, R)
        print("Ma trận QxR: ")
        printMatrix(C)
        print('\n')

    except np.linalg.LinAlgError:
        print("Không thể phân rã QR")

    
A=[[1,1,-1],[0,1,2],[1,1,1]] #phân rã QR được 
A1=[[1,0,1],[-1,1,1],[1,0,1],[-1,1,1]] #Không là họ trực giao -> không thể phân rã QR
A2=[[-2,1,3],[1,0,0],[0,1,0],[0,0,1]] #phân rã QR được 

print("\n---------Test 1---------")
printResolve(A)
print("\n---------Test 2---------")
printResolve(A1)
print("\n---------Test 3---------")
printResolve(A2)
```

- Với hàm QRdecomposition, do thực hiện theo giải thuật Gram - Schmidt nên khi trực giao hóa các vector cột của ma trận A nếu có xuất hiện vector sau khi trực giao hóa là vector 0 thì sẽ kết thúc giải thuật, dẫn đến không thể phân rã ma trận A thành ma trận $Q, R$ và trả về 2 kết quả None. Tuy nhiên, ở hàm linalg.qr của như viện numpy vẫn có thể giải quyết vấn đề này và cho ra kết quả trả về là 2 ma trận $Q_l, R_l$ được phân rã từ ma trận A nhưng để hoàn chỉnh bài làm thì ta sẽ thiết lập những lỗi ngoại lệ sẽ xảy ra đó.

#### Mở rộng: Tìm hiểu và trình bày ứng dụng của QR decomposition
Mở rộng: Ứng dụng của Phân tích QR
Phân tích QR (QR decomposition) là một công cụ trong đại số tuyến tính. Các ứng dụng chính: 

1. Giải hệ phương trình tuyến tính:
QR decomposition giúp giải hệ phương trình tuyến tính một cách hiệu quả. Do ma trận 𝑅 là ma trận tam giác trên, việc giải hệ phương trình dễ dàng hơn so với phương pháp khác.

2. Tính toán trị riêng và vectơ riêng:
Bằng cách dùng phân tích QR, ta tính toán được trị riêng và vectơ riêng của một ma trận. Phương pháp này dùng trong QR algorithm.

3. Tính toán đa thức của ma trận:
QR decomposition cung cấp một cách tính toán đa thức của một ma trận. Điều này có ứng dụng trong nhiều lĩnh vực như lý thuyết đồ thị, xử lý tín hiệu và thống kê.

4. Giảm chiều dữ liệu:
Trong xử lý dữ liệu, phân tích QR có thể được dùng để giảm chiều dữ liệu bằng cách dùng ma trận Q, chúng ta có thể ánh xạ dữ liệu vào một không gian chiều thấp hơn, giúp giảm độ phức tạp tính toán và lưu trữ.

5. Giải các bài toán tối ưu:
QR decomposition được áp dụng rộng rãi trong giải các bài toán tối ưu, bao gồm bài toán tối ưu tuyến tính và phi tuyến tính. Các phương pháp dùng QR decomposition thường giúp tăng hiệu suất và độ chính xác của giải pháp.
