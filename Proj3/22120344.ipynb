#### Gi·∫£i thu·∫≠t ph√¢n r√£ ma tr·∫≠n A     =     QR (Gram ‚Äì Schmidt)

- B∆∞·ªõc 1: Gi·∫£ s·ª≠ m·ªôt ma tr·∫≠n $A$ c√≥ n c·ªôt, m·ªói c·ªôt c·ªßa ma tr·∫≠n l·∫ßn l∆∞·ª£t l√† c√°c vector t∆∞∆°ng ·ª©ng t·ª´ $u_1,u_2,...u_n$
- B∆∞·ªõc 2: ƒê·∫∑t $v_1      =      u_1$, n·∫øu $v_1$ l√† vector 0 th√¨ h·ªç vector ph·ª• thu·ªôc tuy·∫øn t√≠nh v√† k·∫øt th√∫c gi·∫£i thu·∫≠t.
- B∆∞·ªõc 3: ƒê·∫∑t $v_2,v_3,...v_n$ l·∫ßn l∆∞·ª£t v·ªõi $v_m = u_m ‚àí \sum_{i=1}^{m-1}\frac{‚ü®u_m, v_i‚ü©}{‚à•v_i‚à•^{2}}v_i$ (2 <= m <= n). N·∫øu c√≥ b·∫•t k√¨ vector $v_m$ n√†o l√† vector 0 th√¨ h·ªç vector ph·ª• thu·ªôc tuy·∫øn t√≠nh v√† k·∫øt th√∫c gi·∫£i thu·∫≠t. (‚à•v_i‚à• l√† ƒë·ªô d√†i c·ªßa vector).
- B∆∞·ªõc 4: x√¢y d·ª±ng h·ªç c√°c vector tr·ª±c chu·∫©n v·ªõi $q_m = \frac{v_m}{‚à•v_m‚à•}$ (1 <= m <= n). 
- B∆∞·ªõc 5: D·ª±ng ma tr·∫≠n $Q$ g·ªìm n c·ªôt, $Q(q_l,q_2,...,q_n)$
- B∆∞·ªõc 6: D·ª±ng ma tr·∫≠n R l√† m·ªôt ma tr·∫≠n tam gi√°c tr√™n
* V·ªõi
R = $\begin{bmatrix} ‚ü®u_1, q_l‚ü© & ‚ü®u_2, q_l‚ü© & ... & ‚ü®u_n, q_l‚ü© \\
    0 & ‚ü®u_2, q_2‚ü© & ... & ‚ü®u_n, q_2‚ü© \\
    \vdots &\vdots & \ddots &\vdots\\               
    0 & 0 & ... & ‚ü®u_n, q_n‚ü© \end{bmatrix}$

```python
def float_to_fraction(x):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    while abs(decimal - num/denom) > saiSo:
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
    return f"{sign * num}/{denom}"
    

def subtractVector(A,B):
    if(len(A) != len(B)):
        return None
    C = [] 
    for i in range(len(A)):
        C.append(A[i] - B[i])
    return C


def multiVector(c,A):
    D = []
    for i in range(len(A)):
        D.append(c * A[i])
    return D


def printMatrix(A):
    if A is None:
        print('\n')
    else:
        for row in A:
            print([float_to_fraction(elem) for elem in row])
        print('\n')


def multiMatrix(A,B):
    if(len(A[0])!=len(B)):
        print("Kh√¥ng th·ªÉ nh√¢n 2 ma tr·∫≠n\n")
        return None
    C = []
    ans = 0
    for i in range(len(A)):
        tam = []
        for j in range(len(A[0])):
            ans = 0
            for k in range(len(A[0])):
                ans += A[i][k] * B[k][j]
            tam.append(round(ans,3))
        C.append(tam)
    return C



def dotProduct(A,B):
    if(len(A) != len(B)):
        return 0
    tam = 0
    for i in range(len(A)):
        tam += A[i]*B[i]
    return tam




def isVector0(A):
    for i in range(len(A)):
        if A[i] != 0:
            return False
    return True



def norm(A):
    tam = 0.0
    for i in range(len(A)):
        tam += A[i]**2
        
    return tam**(0.5)



def QRdecomposition(A):
    T = []
    tam = []
    for i in range(len(A[0])):
        for j in range(len(A)):
            tam.append(A[j][i]*1);
        T.append(tam)
        tam = []
        
    Q = []
    tam2 = []
    for i in range(len(T[0])):
        tam2.append(T[0][i]*1)
    Q.append(tam2)
    tam2 = []
    
    for i in range(1,len(T)):
        tam2 = T[i]
        for j in range(0,len(Q)):
            tam = multiVector(dotProduct(T[i],Q[j])/(norm(Q[j])**2),Q[j])
            tam2 = subtractVector(tam2,tam)
        if isVector0(tam2) == True:
            print("H·ªá vector ph·ª• thu·ªôc tuy·∫øn t√≠nh n√™n kh√¥ng l√† h·ªç tr·ª±c giao")
            return None,None
        Q.append(tam2)
    
    #tinh truc chuan    
    for i in range(len(Q)):
        chuan1 = norm(Q[i])
        for j in range(len(Q[0])):
            Q[i][j] = Q[i][j]/chuan1
    
    R = []
    for i in range(len(Q)):
        tam = []
        for j in range(len(T)):         
            if j<i:
                tam.append(0)
                continue
            tam.append(dotProduct(Q[i],T[j]))
        R.append(tam)
        
    Qsau = []
    tam = []
    for i in range(len(Q[0])):
        for j in range(len(Q)):
            tam.append(Q[j][i]);
        Qsau.append(tam)
        tam = []
        
    return Qsau,R
  

def printResolve(A):
    if A == None: 
        return
    print("Ma tran A: ")
    printMatrix(A)
    print('\n')

    Q,R = QRdecomposition(A)
    if Q == None or R == None:
        return
    
    print("Ma tr·∫≠n Q: ")
    printMatrix(Q)
    print('\n')

    print("Ma tr·∫≠n R: ")
    printMatrix(R)
    print('\n')

    C=multiMatrix(Q,R)
    print("Ma tr·∫≠n QxR: ")
    printMatrix(C)
    print('\n')
    
A=[[1,1,-1],[0,1,2],[1,1,1]] #ph√¢n r√£ QR ƒë∆∞·ª£c 
A1=[[1,0,1],[-1,1,1],[1,0,1],[-1,1,1]] #Kh√¥ng l√† h·ªç tr·ª±c giao -> kh√¥ng th·ªÉ ph√¢n r√£ QR
A2=[[-2,1,3],[1,0,0],[0,1,0],[0,0,1]] #ph√¢n r√£ QR ƒë∆∞·ª£c 

print("\n---------Test 1---------")
printResolve(A)
print("\n---------Test 2---------")
printResolve(A1)
print("\n---------Test 3---------")
printResolve(A2)
```

### B√†i 2

```python
import numpy as np
np.set_printoptions(suppress=True)

def float_to_fraction(x):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    while abs(decimal - num/denom) > saiSo:
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
    return f"{sign * num}/{denom}"
    
def printMatrix(A):
    if A is None:
        print('\n')
    else:
        for row in A:
            print([float_to_fraction(elem) for elem in row])
        print('\n')



def printResolve(A):
    if A is None: 
        return
    
    print("Ma tr·∫≠n A: ")
    printMatrix(A)
    print('\n')

    if np.linalg.matrix_rank(A) < min(len(A), len(A[0])):
        print("Kh√¥ng th·ªÉ ph√¢n r√£ QR.")
        return

    try:
        Q, R = np.linalg.qr(A)
        print("Ma tr·∫≠n Q: ")
        printMatrix(Q)
        print('\n')

        print("Ma tr·∫≠n R: ")
        printMatrix(R)
        print('\n')

        C = np.dot(Q, R)
        print("Ma tr·∫≠n QxR: ")
        printMatrix(C)
        print('\n')

    except np.linalg.LinAlgError:
        print("Kh√¥ng th·ªÉ ph√¢n r√£ QR")

    
A=[[1,1,-1],[0,1,2],[1,1,1]] #ph√¢n r√£ QR ƒë∆∞·ª£c 
A1=[[1,0,1],[-1,1,1],[1,0,1],[-1,1,1]] #Kh√¥ng l√† h·ªç tr·ª±c giao -> kh√¥ng th·ªÉ ph√¢n r√£ QR
A2=[[-2,1,3],[1,0,0],[0,1,0],[0,0,1]] #ph√¢n r√£ QR ƒë∆∞·ª£c 

print("\n---------Test 1---------")
printResolve(A)
print("\n---------Test 2---------")
printResolve(A1)
print("\n---------Test 3---------")
printResolve(A2)
```

#### M·ªü r·ªông: T√¨m hi·ªÉu h√†m/ph∆∞∆°ng th·ª©c t∆∞∆°ng ·ª©ng c·ªßa c√°c th∆∞ vi·ªán v√† th·ª±c hi·ªán n√≥, so s√°nh k·∫øt qu·∫£

    - Ma tr·∫≠n $Q, R$ nh·∫≠n ƒë∆∞·ª£c t·ª´ h√†m linalg.qr c·ªßa nh∆∞ vi·ªán numpy l√† $Q_l, R_l$

- V·ªõi h√†m QRdecomposition, do th·ª±c hi·ªán theo gi·∫£i thu·∫≠t Gram - Schmidt n√™n khi tr·ª±c giao h√≥a c√°c vector c·ªôt c·ªßa ma tr·∫≠n A n·∫øu c√≥ xu·∫•t hi·ªán vector sau khi tr·ª±c giao h√≥a l√† vector 0 th√¨ s·∫Ω k·∫øt th√∫c gi·∫£i thu·∫≠t, d·∫´n ƒë·∫øn kh√¥ng th·ªÉ ph√¢n r√£ ma tr·∫≠n A th√†nh ma tr·∫≠n $Q, R$ v√† tr·∫£ v·ªÅ 2 k·∫øt qu·∫£ None. Tuy nhi√™n, ·ªü h√†m linalg.qr c·ªßa nh∆∞ vi·ªán numpy v·∫´n c√≥ th·ªÉ gi·∫£i quy·∫øt v·∫•n ƒë·ªÅ n√†y v√† cho ra k·∫øt qu·∫£ tr·∫£ v·ªÅ l√† 2 ma tr·∫≠n $Q_l, R_l$ ƒë∆∞·ª£c ph√¢n r√£ t·ª´ ma tr·∫≠n A nh∆∞ng ƒë·ªÉ ho√†n ch·ªânh b√†i l√†m th√¨ ta s·∫Ω thi·∫øt l·∫≠p nh·ªØng l·ªói ngo·∫°i l·ªá s·∫Ω x·∫£y ra ƒë√≥.

#### M·ªü r·ªông: T√¨m hi·ªÉu v√† tr√¨nh b√†y ·ª©ng d·ª•ng c·ªßa QR decomposition
M·ªü r·ªông: ·ª®ng d·ª•ng c·ªßa Ph√¢n t√≠ch QR
Ph√¢n t√≠ch QR (QR decomposition) l√† m·ªôt c√¥ng c·ª• trong ƒë·∫°i s·ªë tuy·∫øn t√≠nh. C√°c ·ª©ng d·ª•ng ch√≠nh: 

1. Gi·∫£i h·ªá ph∆∞∆°ng tr√¨nh tuy·∫øn t√≠nh:
QR decomposition gi√∫p gi·∫£i h·ªá ph∆∞∆°ng tr√¨nh tuy·∫øn t√≠nh m·ªôt c√°ch hi·ªáu qu·∫£. Do ma tr·∫≠n ùëÖ l√† ma tr·∫≠n tam gi√°c tr√™n, vi·ªác gi·∫£i h·ªá ph∆∞∆°ng tr√¨nh d·ªÖ d√†ng h∆°n so v·ªõi ph∆∞∆°ng ph√°p kh√°c.

3. T√≠nh to√°n ƒëa th·ª©c c·ªßa ma tr·∫≠n:
QR decomposition cung c·∫•p m·ªôt c√°ch t√≠nh to√°n ƒëa th·ª©c c·ªßa m·ªôt ma tr·∫≠n. ƒêi·ªÅu n√†y c√≥ ·ª©ng d·ª•ng trong nhi·ªÅu lƒ©nh v·ª±c nh∆∞ l√Ω thuy·∫øt ƒë·ªì th·ªã, x·ª≠ l√Ω t√≠n hi·ªáu v√† th·ªëng k√™.

5. Gi·∫£i c√°c b√†i to√°n t·ªëi ∆∞u:
QR decomposition ƒë∆∞·ª£c √°p d·ª•ng r·ªông r√£i trong gi·∫£i c√°c b√†i to√°n t·ªëi ∆∞u, bao g·ªìm b√†i to√°n t·ªëi ∆∞u tuy·∫øn t√≠nh v√† phi tuy·∫øn t√≠nh. C√°c ph∆∞∆°ng ph√°p d√πng QR decomposition th∆∞·ªùng gi√∫p tƒÉng hi·ªáu su·∫•t v√† ƒë·ªô ch√≠nh x√°c c·ªßa gi·∫£i ph√°p.

### M√¥ t·∫£ c√°c h√†m v√† √Ω t∆∞·ªüng th·ª±c hi·ªán ch∆∞∆°ng tr√¨nh kh√¥ng d√πng th∆∞ vi·ªán

- Duy·ªát qua t·ª´ng c·ªôt a c·ªßa A. ƒê·ªëi v·ªõi m·ªói c·ªôt a, th·ª±c hi·ªán c√°c b∆∞·ªõc:

    * Ki·ªÉm tra xem u c√≥ ph·∫£i l√† vector kh√¥ng hay kh√¥ng. N·∫øu u l√† vector kh√¥ng (t·ª©c l√† c√°c c·ªôt c·ªßa A ph·ª• thu·ªôc tuy·∫øn t√≠nh), in th√¥ng b√°o l·ªói v√† tr·∫£ v·ªÅ None, None.
    * Chu·∫©n h√≥a u b·∫±ng c√°ch chia u cho chu·∫©n c·ªßa u. Th√™m u v√†o Q.

- Sau khi duy·ªát qua t·∫•t c·∫£ c√°c c·ªôt c·ªßa A, tr·∫£ v·ªÅ Q v√† R.
##### L∆∞u √Ω: Thu·∫≠t to√°n Gram-Schmidt ch·ªâ ho·∫°t ƒë·ªông ƒë√∫ng n·∫øu c√°c c·ªôt c·ªßa A ƒë·ªôc l·∫≠p tuy·∫øn t√≠nh. N·∫øu kh√¥ng, h√†m QRdecomposition(A) s·∫Ω kh√¥ng th·ªÉ t·∫°o ra k·∫øt qu·∫£ h·ª£p l·ªá v√† s·∫Ω tr·∫£ v·ªÅ None, None.

### C√°c h√†m trong ch∆∞∆°ng tr√¨nh d√πng th∆∞ vi·ªán:

* printResolve(A): In ma tr·∫≠n A v√† k·∫øt qu·∫£ c·ªßa ph√¢n r√£ QR (n·∫øu c√≥ th·ªÉ). N·∫øu A l√† None, h√†m s·∫Ω kh√¥ng l√†m g√¨.
