### Giải thuật tìm ma trận nghịch đảo A<sup>-1</sup>
- Bằng phép khử Gauss-Jordan sẽ sử dụng các phép biến đổi ma trận để chuyển ma trận A thành ma trận đơn vị I. Sau đó, áp dụng cùng các biến đổi đó vào ma trận đơn vị I, ta thu được ma trận nghịch đảo A<sup>-1</sup>
- Giải thuật được thực hiện như sau:
    + Bước 1: Tạo ma trận bổ sung A' bằng cách ghép ma trận đơn vị I vào bên phải của ma trận A.
    + Bước 2: Áp dụng phép khử Gauss - Jordan vào ma trận bổ sung A' để chuyển ma trận A thành ma trận bậc thang. Lưu ý rằng phép khử Gauss-Jordan được áp dụng đồng thời vào cả ma trận A và I.
    + Bước 3: Áp dụng phép biến đổi đến ma trận bậc thang A' để chuyển ma trận A thành ma trận đơn vị I.
    + Bước 4: Ma trận đơn vị I' mới thu được sau các phép biến đổi trong bước 2 là ma trận nghịch đảo của ma trận A. Tức là I' chính là ma trận A<sup>-1</sup>


```python
def float_to_fraction(x):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    while abs(decimal - num/denom) > saiSo:
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
    return f"{sign * num}/{denom}"
    

def print_matrix(A):
    if A is None:
        print('\n')
    else:
        for row in A:
            print([float_to_fraction(elem) for elem in row])
        print('\n')

def create_identity_matrix(n):
    A = []
    for i in range(0, n):
        temp = []
        for j in range(0, n):
            if i == j:
                temp.append(1)
            else:
                temp.append(0)            
        A.append(temp)
    return A

def inverse(A):
    # Kiểm tra nếu ma trận A input không hợp lệ
    for i in range(len(A)):
        if len(A) != len(A[i]):
            print("Ma trận input không vuông, không thể tìm ma trận nghịch đảo!")
            return None
    
    # Tạo ma trận đơn vị In
    B = create_identity_matrix(len(A))
        
    flag = True
    j = 0
    
    # Duyệt qua từng cột của ma trận A
    for i in range(len(A[0])):
        k = j
        flag = False
        
        # Duyệt qua từng dòng của cột đó để kiểm tra cột có giá trị khác 0 hay không
        for l in range(j, len(A)):
            if A[l][i] != 0:
                flag = True
                
        if flag:
            if A[k][i] == 0:
                for z in range(k + 1, len(A)):
                    if A[z][i] != 0:
                        A[k], A[z] = A[z], A[k]
                        B[k], B[z] = B[z], B[k]
                        break
            
            d = (1 / A[k][i])
            for m in range(len(A[0])):
                A[k][m] *= d
                B[k][m] *= d
            
            for k1 in range(len(A)):
                if k1 == i:
                    continue
                d = -A[k1][i] / A[k][i]
                for m in range(len(A[0])):
                    A[k1][m] += d * A[k][m]
                    B[k1][m] += d * B[k][m]
            j += 1
        else:
            print("Ma trận không khả nghịch")
            return None
    return B

# Trường hợp khả nghịch
A = [[0, 1, 2], [1, 2, 1], [1, -1, 3]] 
print("Ma trận A: ")
print_matrix(A)
print("Ma trận nghịch đảo của A: ")
A_inv = inverse(A)
print_matrix(A_inv)

# Trường hợp không khả nghịch
B = [[1, 3, -4], [1, 5, -1], [3, 13, -6]] 
print("Ma trận B: ")
print_matrix(B)
print("Ma trận nghịch đảo của B: ")
B_inv = inverse(B)
print_matrix(B_inv)

# Trường hợp không khả nghịch
C = [[1, 3, -4], [1, 5, -1], [3, 13, -6], [1, 5, 9]] 
print("Ma trận C: ")
print_matrix(C)
print("Ma trận nghịch đảo của C: ")
C_inv = inverse(C)
print_matrix(C_inv)
```

```python
## Mở rộng: Tìm hiểu hàm/ phương thức tương ứng của các thư viện và thực hiện nó, so sánh kết quả
import numpy as np
def float_to_fraction(x, max_denom=10000, max_iterations=10000):
    if x % 1 == 0:
        return str(int(x))
    saiSo = 1.0E-6
    num, denom = 1, 1
    decimal, sign = abs(x), -1 if x < 0 else 1
    best_num, best_denom, best_diff = num, denom, abs(decimal - num/denom)
    iterations = 0
    while denom <= max_denom and iterations < max_iterations:
        if abs(decimal - num/denom) < best_diff:
            best_diff = abs(decimal - num/denom)
            best_num, best_denom = num, denom
        if decimal < num/denom:
            denom += 1
        else:
            num += 1
        iterations += 1
    return f"{sign * best_num}/{best_denom}"
    
#Trường hợp khả nghịch
A = np.array([[0, 1, 2],
              [1, 2, 1],
              [1, -1, 3]])


#Trường hợp không khả nghịch
B = np.array([[1, 3, -4],
              [1, 5, -1],
              [3, 13, -6]])

#Trường hợp không khả nghịch
C = np.array([[1, 3, -4],
              [1, 5, -1],
              [3, 13, -6],
              [1, 5, 9]])
def compute_matrix_inverse(matrix):
    try:
        inverse_matrix = np.linalg.inv(matrix)
        print("Ma tran nghich dao cua ma tran:")
        for row in inverse_matrix:
            print([float_to_fraction(x) for x in row])
    except np.linalg.LinAlgError:
        print("Ma tran khong kha nghich (khong co ma tran nghich dao).")
print("Ma tran A:")
print(A)
compute_matrix_inverse(A)

print("\nMa tran B:")
print(B)
compute_matrix_inverse(B)

print("\nMa tran C:")
print(C)
compute_matrix_inverse(C)
```

- Đối với ma trận không khả nghịch:
    + Hàm inverse tự định nghĩa sẽ thông báo ma trận không khả nghịch
    + Hàm linalg.inv của thư viện numpy cho ra một ma trận kết quả với các phần tử đều có giá trị vô cùng lớn/vô cùng bé
- Đối với ma trận không vuông:
    + Hàm inverse tự định nghĩa sẽ thông báo ma trận không vuông, không thể tìm ma trận nghịch đảo
    + Hàm linalg.inv của thư viện numpy sẽ thông báo lỗi chương trình
